```python
def loadVolume(path):
    lh = path + '.lh.func.gii'
    rh = path + '.rh.func.gii'
    lh_data = nib.load(lh).agg_data()
    rh_data = nib.load(rh).agg_data()
    arr = np.concatenate((lh_data, rh_data))
    return(arr)

volume = loadVolume('../Figure5Data/PropMaxValue/Volume_Human/func.gii.diff/M.12-6')
#area = loadVolume('../Figure3Data/MacaquePeaksInHuman/aparc/DifferenceMaps/volumes/M.Ranked.Area.nii.gz')
#thickness = loadVolume('../Figure3Data/MacaquePeaksInHuman/aparc/DifferenceMaps/volumes/M.Ranked.Thickness.nii.gz')
```
